```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
```

```python

res = requests.get("http://web.vscinemas.com.tw/film/detail.aspx")

soup = BeautifulSoup(res.text, 'html.parser')

print(soup)
```

```python
links = soup.find_all('a')
url = 'detail.aspx?id='
df = pd.Series()
movieurl = pd.Series()
for link in links:
    li =link.get('href')
    df = df.append(pd.Series(li)).reset_index(drop=True)

for d in df:
    if url in d:
        movieurl = movieurl.append(pd.Series(d)).reset_index(drop=True)
movieurl = movieurl.drop_duplicates().reset_index(drop=True)
movieurl.to_csv('murl.csv')

        
```
